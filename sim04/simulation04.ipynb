In this exercise, a molecular dynamics simulation will be presented. The system under study is a Lennard-Jones model for Argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu) in the following conditions:
1. solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$).

The units of measure will be reasonabily converted in the IS units when the plots will be shown.
The pressure in a system, subjectd to a Lennard-Jones potential, is extimated thanks to the virial expansion as follows:
$$ P = \rho k_B T + \frac{1}{3V} \left\langle \sum_{i=1}^{N-1} \sum_{j\,(>i)}^N 48\epsilon \left[ \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - 
\frac{1}{2} \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] \right\rangle $$
The datas that will be showcased, have been obtained thanks to a reasonable equilibration of the system.
The temperature, starting from a cristalline configuration for example, undergoes a strong oscillation during the first time steps of the simulation. This is cused by the fact that at the beginning the system is in a perfectly ordered state (cristalline solid for example), in which the value of the potential energy is lower respect to the one in a more disordered state (non perfect solid). During the simulation, the system equlibrates, augmenting the value of its potential enenrgy. By sensibly modifing the initial value of the temperature, by increasing or decresing it, with respect to the desired value, and after restarting the simulation, by changing the value of the boolean variable restart, usually 30000 MC steps, the system, reaches the equilibrium, and the variuos phase simulations can be performed.

#### TD quantities: solid phase 

```python
#constants
temperature = 120

kb = 1.30*10**(-23)

energy = temperature*kb
sigma = 0.34
pressure = energy/(sigma**3)

plt.figure(figsize = (14,27))
plt.subplot(5,1,1)
plt.title('mean kinetic energy')
N, ekin, err_ekin = np.loadtxt("MD/solid/ave_ekin.dat", usecols = (0,2,3), unpack=True)
plt.errorbar(np.arange(100), ekin*energy, yerr=err_ekin*energy,fmt = "bo", capsize = 5, elinewidth = 1, label = "mean kinetic energy")
plt.xlabel('number of blocks')
plt.grid('True')
plt.legend()

plt.subplot(5,1,2)
plt.title('mean potential energy')
N, epot, err_epot = np.loadtxt("MD/solid/ave_epot.dat", usecols = (0,2,3), unpack=True)
plt.errorbar(np.arange(100), epot*energy, yerr=err_epot*energy, fmt = "bo", capsize = 5, elinewidth = 1, label = "mean potential energy")
plt.xlabel('Number of blocks')
plt.grid('True')
plt.legend()

plt.subplot(5,1,3)
plt.title('mean temperature')
N, temp, err_temp = np.loadtxt("MD/solid/ave_temp.dat", usecols = (0,2,3), unpack=True)
plt.errorbar(np.arange(100), temp*120, yerr=err_temp*120, fmt = "bo", capsize = 5, elinewidth = 1, label = "mean temperature")
plt.axhline(y=0.8*120, xmin=0, xmax=100, color = 'red', label='reference temperature')
plt.xlabel('Number of blocks')
plt.grid('True')
plt.legend()

                   
plt.subplot(5,1,4)
plt.title('mean total energy')
N, etot, err_etot = np.loadtxt("MD/solid/ave_etot.dat", usecols = (0,2,3), unpack=True)
plt.errorbar(np.arange(100), etot*energy, yerr=err_etot*energy, fmt = "bo", capsize = 5, elinewidth = 1, label = "mean total energy")
plt.xlabel('Number of blocks')
plt.grid('True')
plt.legend()
                   

                   
plt.subplot(5,1,5)
plt.title('mean pressure')
N, pres, err_pres = np.loadtxt("MD/solid/ave_pres.dat", usecols = (0,2,3), unpack=True)
plt.errorbar(np.arange(100), pres*pressure, yerr=err_pres*pressure, fmt = "bo", capsize = 5, elinewidth = 1, label = "mean pressure")
plt.xlabel('Number of blocks')
plt.grid('True')
plt.legend()
```
