### <span style="color:orange"> The Traveling Salesman Problem</span>

Here a will show a simple genetic agorithm to solve the TSP with 34 cities, organized on a circumference and in a square.
This are the main operators:
- check functions: they check if, for each elements of a given population, every city is visited at least but no more than once;
- mutation operator: it makes different types of permutations between cities in an element of the population; 
- selection operator: it selects the best elements of each population, according to the cost function selected, $L^{2}$ norm
- cross over operator: given a father and a mother selected from the population, produces two children which will be added to the new population. The mother and the father will be selected from the actual population by the selction operator.

The mutations (4 types) and the cross over operations are selected with the following probabilities:
$$
p = \lbrace 0.08, 0.08, 0.08, 0.08, 0.64 \rbrace
$$
Where the highest probability corresponds to the **crossover** operation.

```python
import numpy as np
import matplotlib.pyplot as plt

L2_circle = np.load('L2_circle1.npy')
Worst_circle = np.load('Worst_circle1.npy') #temporary
Best_circle =  np.load('Best_circle1.npy') # best path




fig,(ax, ax1)=plt.subplots(nrows= 1, ncols=2, figsize=(16,7))
ax.set_title('Initial path', fontsize = 20)
ax.scatter(Worst_circle[:,1],Worst_circle[:,2])
z=np.zeros((35,3))
z[0:34,:]= Worst_circle
z[34,:]=Worst_circle[0,:]
ax.plot(z[:,1],z[:,2])
for i in range(34):
    ax.text(x=Worst_circle[i,1], y=Worst_circle[i,2], s=np.array2string(Worst_circle[i,0]))

ax.grid('True')

ax1.set_title('Final path on a circle: 500 generations', fontsize = 20)
ax1.scatter(Best_circle[:,1,0],Best_circle[:,2,0])
z=np.zeros((35,3))
z[0:34,:]= Best_circle[:,:,0]
z[34,:]=Best_circle[0,:,0]
ax1.plot(z[:,1],z[:,2],color='red')
for j in range(34):
    ax1.text(x=Best_circle[j,1,0], y=Best_circle[j,2,0], s=np.array2string(Best_circle[j,0,0]))

ax1.grid('True')


L2_square = np.load('L2_square1.npy')
Worst_square = np.load('Worst_squaare1.npy') #temporary
Best_square =  np.load('Best_square1.npy') # best path




fig,(ax2, ax3)=plt.subplots(nrows= 1, ncols=2, figsize=(16,7))
ax2.set_title('Initial path', fontsize = 20)
ax2.scatter(Worst_square[:,1],Worst_square[:,2])
z=np.zeros((35,3))
z[0:34,:]= Worst_square
z[34,:]=Worst_square[0,:]
ax2.plot(z[:,1],z[:,2])
for i in range(34):
    ax2.text(x=Worst_square[i,1], y=Worst_square[i,2], s=np.array2string(Worst_square[i,0]))
ax2.grid('True')

ax3.set_title('Final path on a square: 1000 generations', fontsize = 20)
ax3.scatter(Best_square[:,1,0],Best_square[:,2,0])

z=np.zeros((35,3))
z[0:34,:]= Best_square[:,:,0]
z[34,:]=Best_square[0,:,0]
ax3.plot(z[:,1],z[:,2],color='red')
#ax3.plot(Best_square[:,1,0],Best_square[:,2,0],color='red')
for j in range(34):
    ax3.text(x=Best_square[j,1,0], y=Best_square[j,2,0], s=np.array2string(Best_square[j,0,0]))

ax3.grid('True')
```

### Analysis of the $L^{2}$ norm, seen as a cost function
This is the behaviour of the $L^{2}$, which is used as a cost function in both GA searches and it is used to define the fittest elements of the population, thanks to the selection operator.
