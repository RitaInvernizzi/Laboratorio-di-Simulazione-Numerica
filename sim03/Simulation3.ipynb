The following graphs depict the behaviours of the European Call-option price and the European Put-option price, obtiained by  directly samplig a geometric brownian motion distribution of the final asset price and by sampling the discretized $GBM(r, \sigma^{2})$ path of the $S(T)$. The $GBM(r, \sigma^{2})$ path is discretized by dividig the time span $[0,T]$ in 100 subintervals.

```python
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt

y = np.loadtxt('SCallC.txt', delimiter=',', unpack=True)
errs = np.loadtxt('SECallC.txt', delimiter=',', unpack=True)
x=np.arange(100)
comp=14.975790778311286*np.ones(100)
z = np.loadtxt('SPutC.txt', delimiter=',', unpack=True)
errsz = np.loadtxt('SEPutC.txt', delimiter=',', unpack=True)
comp1=14.975790778311286*np.ones(100)
comp2=5.4595325819072364*np.ones(100)
fig = plt.figure(figsize=(10,7)) 
plt.errorbar(x,y,yerr=errs, label='MC Simulation Call')
plt.plot(x,comp1,'--', label= 'Analytical Result Call')
plt.errorbar(x,z,yerr=errsz, label='MC Simulation Put')
plt.plot(x,comp2,'--', label= 'Analytical Result Put')
plt.title('European Call & Put Option: Direct Sampling',fontsize=20)
plt.ylim(-1,21)
plt.xlabel('Number of Blocks', fontsize=15)
#plt.ylabel('Call Direct Sampling', fontsize=15)
plt.legend(loc="upper right", fontsize=13,frameon=True)
plt.grid(True)
plt.show()
```
